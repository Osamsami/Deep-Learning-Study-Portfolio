## Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Artificial Neuron

```python
x = np.array([1.5, 2.0])
w = np.array([0.4, 0.6])
b = 0.2

z = np.dot(x, w) + b
print("Weighted Sum:", z)
```

## Activation Function

```python
def step_function(z):
    return 1 if z >= 0 else 0

output = step_function(z)
print("Output:", output)
```

## Dataset (AND Gate)

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,0,0,1])
```

## Initialize Parameters

```python
weights = np.random.rand(2)
bias = np.random.rand()
lr = 0.1
```

## Prediction Function

```python
def predict(x):
    z = np.dot(x, weights) + bias
    return 1 if z >= 0 else 0
```

## Training Loop

```python
losses = []

for epoch in range(10):
    total_error = 0

    for i in range(len(X)):
        y_pred = predict(X[i])
        error = y[i] - y_pred

        weights[:] = weights + lr * error * X[i]
        bias += lr * error

        total_error += abs(error)

    losses.append(total_error)
    print(f"Epoch {epoch+1}, Error: {total_error}")
```

## Loss Graph

```python
plt.plot(losses)
plt.title("Training Error")

save_path = base_path + "/Visuals/loss_plot.png"
plt.savefig(save_path)
plt.show()
```

## Testing

```python
for i in range(len(X)):
    print(X[i], "->", predict(X[i]))
```

## Decision Boundary

```python
x1 = np.linspace(-0.2, 1.2, 10)
x2 = -(weights[0]*x1 + bias) / weights[1]

plt.scatter(X[:,0], X[:,1], c=y)
plt.plot(x1, x2)

plt.title("Decision Boundary")

save_path = base_path + "/Visuals/decision_boundary.png"
plt.savefig(save_path)
plt.show()
```
